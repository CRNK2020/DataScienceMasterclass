```python
chipo['item_price']
```

```python
#apply function
chipo['item_price_float'] = chipo['item_price'].apply(lambda x: float(x[1:]))
```

### Step 4. How many products cost more than $10.00?

```python
chipo
```

```python
print("Total products cost more than $10.00 in order id level :",len(chipo[chipo['item_price_float']>10.00]['item_name']))
```

```python

print("Total products cost more than $10.00 is :",chipo[chipo['item_price_float']>10.00]['item_name'].nunique())
```

```python
set(chipo[chipo['item_price_float']>10.00]['item_name'])
print("Total products cost more than $10.00 is :",len(set(chipo[chipo['item_price_float']>10.00]['item_name'])))
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo['price_of_one_quantity'] = chipo['item_price_float']/chipo['quantity']
```

```python
chipo
```

```python
result = chipo[['item_name','price_of_one_quantity']]
```
